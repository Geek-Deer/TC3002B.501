```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from fitter import Fitter
```

```python
df = pd.read_csv('./sample_data/sleep_data.csv')
```

```python
print(df.head())
print(df.columns)


column = 'Hours_of_Sleep'

plt.figure(figsize=(10, 6))
sns.histplot(df[column], kde=True, bins=20, color='skyblue', edgecolor='black')
plt.title(f'Histograma de {column}')
plt.xlabel(column)
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
f = Fitter(df[column], distributions=['norm', 'gamma', 'expon', 'lognorm', 'uniform'])
```

```python
# Fit the distributions
f.fit()

# Get the best fitting distribution
best_dist = f.get_best(method='sumsquare_error')

# Print
print(f"Best fitting distribution: {best_dist}")

# Plot
f.summary()
plt.title(f"Distribution Fitting for {column}")
plt.show()
```
